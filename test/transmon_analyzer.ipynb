```python
import numpy as np
alpha = 0.01 * 2 * np.pi 
gamma = 1.0                 

def lorentzian_spectrum(omega, T):
    return (2 * np.pi * alpha) / (omega**2 + gamma**2)

eigenenergies, evecs = testsc.scEigensys(transmon1)
#计算特征值和相应的特征向量使用scipy.linalg.eigh.。返回两个numpy数组，分别包含特征值和特征向量

noise_op = np.zeros((evecs.shape[0], evecs.shape[0]))  
noise_op[0, 1] = 1.0  
noise_op[1, 0] = 1.0  

testsc.scT1(transmon1, 0, 1, noise_op, lorentzian_spectrum)
#使用费米黄金法则计算过渡时间（或速率），由于噪声通道具有谱密度spectral_density和系统噪声算子noise_op
```
